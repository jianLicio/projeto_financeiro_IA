```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

df_treinamento = pd.read_csv('../arquivos_csv/treinamento3.csv', low_memory=False)
df_teste = pd.read_csv('../arquivos_csv/teste3.csv', low_memory=False)
```

### Removendo valores negativos e aplicando a moda

```python
for column in df_treinamento.columns:
    mode_value = df_treinamento[df_treinamento[column] != -1][column].mode()[0]
    
    df_treinamento[column] = df_treinamento[column].replace(-1, mode_value)

remaining_neg_ones = (df_treinamento == -1).sum().sum()

print(f"Quantidade de valores -1 restantes em todo o DataFrame: {remaining_neg_ones}")
print("Substituição realizada com sucesso para todos os atributos.")
```

```python
for column in df_teste.columns:
    mode_value = df_teste[df_teste[column] != -1][column].mode()[0]
    
    df_teste[column] = df_teste[column].replace(-1, mode_value)

remaining_neg_ones = (df_teste == -1).sum().sum()

print(f"Quantidade de valores -1 restantes em todo o DataFrame: {remaining_neg_ones}")
print("Substituição realizada com sucesso para todos os atributos.")
```

### Idade

```python
inertia = []
K_range = range(2, 25)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_treinamento[['Age']].dropna())
    inertia.append(kmeans.inertia_)

plt.plot(K_range, inertia, marker='o')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('Inércia')
plt.title('Método do Cotovelo para Determinar K')
plt.show()
```

```python
df = df_treinamento[['Age']].dropna()

k = 6
kmeans = KMeans(n_clusters=k, random_state=0)

df['idade'] = kmeans.fit_predict(df)

df_treinamento['idade'] = np.nan
df_treinamento.loc[df.index, 'idade'] = df['idade']

print("Centros dos clusters:", kmeans.cluster_centers_)
df_treinamento.drop(columns=['Age'], inplace=True)
```

```python
df = df_teste[['Age']].dropna()

k = 6
kmeans = KMeans(n_clusters=k, random_state=0)

df['idade'] = kmeans.fit_predict(df)

df_teste['idade'] = np.nan
df_teste.loc[df.index, 'idade'] = df['idade']

print("Centros dos clusters:", kmeans.cluster_centers_)
df_teste.drop(columns=['Age'], inplace=True)
```
