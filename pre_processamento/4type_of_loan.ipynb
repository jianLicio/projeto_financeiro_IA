```python
import pandas as pd
import numpy as np

df_treinamento = pd.read_csv('../arquivos_csv/treinamento2.csv', low_memory=False)
df_teste = pd.read_csv('../arquivos_csv/teste2.csv', low_memory=False)
```

```python

atributos = [
    'Annual_Income',
    'Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan',
    'Type_of_Loan', 'Delay_from_due_date', 'Num_of_Delayed_Payment',
    'Changed_Credit_Limit', 'Num_Credit_Inquiries', 'Credit_Mix',
    'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Credit_History_Age',
    'Payment_of_Min_Amount', 'Total_EMI_per_month', 'Amount_invested_monthly',
    'Payment_Behaviour', 'Monthly_Balance', 'Credit_Score'
]

for atributo in atributos:
    unique_values = df_treinamento[atributo].unique()
    print(f"Valores únicos no atributo '{atributo}':")
    print(unique_values)
    print("\n")
    
```

### Num_of_Loan

```python
df_treinamento['Num_of_Loan'] = df_treinamento['Num_of_Loan'].str.replace('_', '')
df_treinamento['Num_of_Loan'] = pd.to_numeric(df_treinamento['Num_of_Loan'], errors='coerce')
```

```python
df_teste['Num_of_Loan'] = df_teste['Num_of_Loan'].str.replace('_', '')
df_teste['Num_of_Loan'] = pd.to_numeric(df_teste['Num_of_Loan'], errors='coerce')
```

```python
num_negativos = (df_treinamento['Num_of_Loan'] < 0).sum()

print(f"Total de instâncias com 'Num_of_Loan' negativos: {num_negativos}")
```

```python
num_negativos = (df_teste['Num_of_Loan'] < 0).sum()

print(f"Total de instâncias com 'Num_of_Loan' negativos: {num_negativos}")
```

```python
df_treinamento.loc[df_treinamento['Num_of_Loan'] < 0, 'Num_of_Loan'] = np.nan
```

```python
df_teste.loc[df_teste['Num_of_Loan'] < 0, 'Num_of_Loan'] = np.nan
```

```python
maior_valor = df_treinamento['Num_of_Loan'].max()

# Exibir o maior valor
print(f"O maior valor em 'Num_of_Loan' é: {maior_valor}")
```

```python
maior_valor = df_teste['Num_of_Loan'].max()

# Exibir o maior valor
print(f"O maior valor em 'Num_of_Loan' é: {maior_valor}")
```

### Type_of_Loan

```python
def ordenar_tipos_emprestimos(valor):
    if isinstance(valor, str):
        # Quebrar em partes por vírgula, remover espaços, remover duplicatas e ordenar
        partes = [parte.strip() for parte in valor.split(',')]
        partes_unicas = sorted(set(partes))  # Remover duplicatas e ordenar
        return ', '.join(partes_unicas)
    return valor

df_treinamento['Type_of_Loan'] = df_treinamento['Type_of_Loan'].apply(ordenar_tipos_emprestimos)

valores_unicos = df_treinamento['Type_of_Loan'].unique()

valores_filtrados = [valor for valor in valores_unicos if isinstance(valor, str)]

valores_ordenados = sorted(valores_filtrados)

print("Valores únicos no atributo 'Type_of_Loan' em ordem alfabética:")
for valor in valores_ordenados:
    print(valor)
```

```python
def ordenar_tipos_emprestimos(valor):
    if isinstance(valor, str):
        # Quebrar em partes por vírgula, remover espaços, remover duplicatas e ordenar
        partes = [parte.strip() for parte in valor.split(',')]
        partes_unicas = sorted(set(partes))  # Remover duplicatas e ordenar
        return ', '.join(partes_unicas)
    return valor

df_teste['Type_of_Loan'] = df_teste['Type_of_Loan'].apply(ordenar_tipos_emprestimos)

valores_unicos = df_teste['Type_of_Loan'].unique()

valores_filtrados = [valor for valor in valores_unicos if isinstance(valor, str)]

valores_ordenados = sorted(valores_filtrados)

print("Valores únicos no atributo 'Type_of_Loan' em ordem alfabética:")
for valor in valores_ordenados:
    print(valor)
```

```python
valores_numericos = df_treinamento['Type_of_Loan'].apply(lambda x: isinstance(x, (int, float)))
valores_numericos_nao_nulos = df_treinamento.loc[valores_numericos & df_treinamento['Type_of_Loan'].notnull(), 'Type_of_Loan']

# Exibir os valores encontrados
print(valores_numericos_nao_nulos)
```
